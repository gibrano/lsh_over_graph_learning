```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Parámetros
n_nodes = [5, 5, 10, 15]   # cantidad de nodos
radii = [0.1, 0.2, 0.1, 0.2]   # radios de conexión
seeds = [1, 2, 3, 4]           # semillas para reproducibilidad
```

```python




fig, axes = plt.subplots(2, 2, figsize=(10, 10))

for ax, n, r, seed in zip(axes.flatten(), n_nodes, radii, seeds):
    # Crear grafo geométrico aleatorio
    G = nx.random_geometric_graph(n, r, seed=seed)

    # Obtener posiciones (ya vienen asignadas en G.graph['pos'])
    pos = nx.get_node_attributes(G, "pos")

    # Dibujar nodos y aristas
    nx.draw_networkx_edges(G, pos, ax=ax, alpha=0.4)
    nx.draw_networkx_nodes(G, pos, ax=ax, node_size=50, )

    ax.set_title(f"RGG con n={n}, r={r}", fontsize=12)
    ax.set_axis_off()

plt.tight_layout()
plt.show()
```

```python
# Parámetros
n_nodes = [5, 5, 10, 10]   # cantidad de nodos
radii = [0.1, 0.2, 0.1, 0.2]   # radios de conexión
seeds = [1, 2, 3, 4]           # semillas para reproducibilidad

fig, axes = plt.subplots(2, 2, figsize=(12, 12))

for ax, n, r, seed in zip(axes.flatten(), n_nodes, radii, seeds):
    # Crear grafo geométrico aleatorio
    G = nx.random_geometric_graph(n, r, seed=seed)

    # Obtener posiciones (ya vienen asignadas en G.graph['pos'])
    pos = nx.get_node_attributes(G, "pos")

    # Dibujar aristas
    nx.draw_networkx_edges(G, pos, ax=ax, alpha=0.4)

    # Dibujar nodos
    nx.draw_networkx_nodes(G, pos, ax=ax, node_size=80, node_color="steelblue")

    # Agregar etiquetas de nodos
    nx.draw_networkx_labels(G, pos, ax=ax, font_size=7, font_color="black")

    ax.set_title(f"RGG con n={n}, r={r}", fontsize=12)
    #ax.set_axis_off()

#plt.tight_layout()
plt.show()
```
